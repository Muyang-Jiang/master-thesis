```python
import geopandas as gpd
import pandas as pd
import numpy as np
import os
import csv
from tqdm import tqdm

from shapely.geometry import Point
import matplotlib.pyplot as plt
from shapely.geometry import Polygon
import folium
```

```python
poi_path = r"/data-sat-img/postprocessed/mujiang/POI/全国POI数据/"
save_path = "/data-sat-img/postprocessed/mujiang/master-thesis/poi_summary/"
city_path = r"/data-sat-img/postprocessed/mujiang/POI/cities/"
block_path = "/data-sat-img/postprocessed/mujiang/master-thesis/block/"
geojson_path = "/data-sat-img/postprocessed/mujiang/master-thesis/poi_geojson/"
```

```python
geomaster_path = "/data-sat-img/postprocessed/mujiang/geobound_master/"
geomaster = pd.read_csv(geomaster_path + "lookup_table.csv")

geomaster_selected = geomaster[["MID_m","地级","省级"]]
```

```python
city  = "深圳市"
MID_city = geomaster_selected.loc[geomaster_selected["地级"] == city]["MID_m"].values.min()
MID_city
```

## read data

```python
year_list = list(range(2018,2023))
poi_df_list = []

for year in year_list:
    file_path = os.path.join(city_path,f"{MID_city}_{city}",f"{year}_{MID_city}_{city}.csv")
    poi_df_list.append(pd.read_csv(os.path.join(file_path),encoding = "utf-8-sig"))
```

```python
year_list
```

```python
poi_df_list[0].head(2)
```

## attach the class1 label

```python
class1_map = pd.read_csv(save_path + "class1_map.csv")
```

```python
poi_df_merged_list = []

for poi_df in poi_df_list: 
    poi_df_merged = pd.merge(poi_df,class1_map,on = "class1",how = "left")
    poi_df_merged_list.append(poi_df_merged)
```

```python
poi_df_merged_list[0].head(2)
```

### sptatial join

```python
#read block data
block_gdf = gpd.read_file(block_path + "shenzhen_block_v1.shp")
block_gdf["block_id"] = block_gdf.index
```

```python
block_gdf.head()
```

```python
joined_gdf_list = []
for poi_df_merged in poi_df_merged_list:
    # turn into geo dataframe
    poi_gdf_merged = gpd.GeoDataFrame(poi_df_merged, geometry=gpd.points_from_xy(poi_df_merged.lon, poi_df_merged.lat))
    poi_gdf_merged = poi_gdf_merged.set_crs(block_gdf.crs, allow_override=True)
    
    joined_gdf = gpd.sjoin(block_gdf,poi_gdf_merged, how="inner", op='intersects')
    joined_gdf_list.append(joined_gdf)
    
```

```python
joined_gdf_list[0].head(2)
```

### find poi that appear most frequently in a block

```python
i = 0
plt.figure(figsize=(8,5))

for joined_gdf in joined_gdf_list:
    block_top_class1 = joined_gdf.groupby('block_id')['class1_label'].agg(lambda x: x.mode()[0]).reset_index(name='Mode')
    #print(year_list[i])
    
    #print(block_top_class1.value_counts(["Mode"]).sort_index())
    
    frequency = block_top_class1.value_counts(["Mode"]).sort_index()
    frequency.index = frequency.index.map(lambda x: int(', '.join(map(str, x))))
    plt.plot(frequency.index, frequency.values, marker='o', linestyle='-', markersize=8,label = year_list[i])

    i+=1
plt.legend()
plt.show()
```

## analysis on filtered data

```python
class1_label_filter_list = [-1,0,10]
```

```python
joined_gdf_list = []
for poi_df_merged in poi_df_merged_list:
    # turn into geo dataframe
    poi_gdf_merged = gpd.GeoDataFrame(poi_df_merged, geometry=gpd.points_from_xy(poi_df_merged.lon, poi_df_merged.lat))
    poi_gdf_merged = poi_gdf_merged.set_crs(block_gdf.crs, allow_override=True)
    poi_gdf_merged =  poi_gdf_merged[~poi_gdf_merged['class1_label'].isin(class1_label_filter_list)]
    
    joined_gdf = gpd.sjoin(poi_gdf_merged,block_gdf, how="inner", op='intersects')
    joined_gdf_list.append(joined_gdf)
```

```python
i = 0
plt.figure(figsize=(8,5))
block_top_class1_list = []


for joined_gdf in joined_gdf_list:
    block_top_class1 = joined_gdf.groupby('block_id')['class1_label'].agg(lambda x: x.mode()[0]).reset_index(name='Mode')
    block_top_class1_list.append(block_top_class1)
    #print(year_list[i])
    
    #print(block_top_class1.value_counts(["Mode"]).sort_index())
    
    frequency = block_top_class1.value_counts(["Mode"]).sort_index()
    frequency.index = frequency.index.map(lambda x: int(', '.join(map(str, x))))
    plt.plot(frequency.index, frequency.values, marker='o', linestyle='-', markersize=8,label = year_list[i])

    i+=1
plt.legend()
plt.show()
```

## spatially plot the block type

```python
block_gdf_classed_list = []
for i in range(len(year_list)):
    block_gdf_classed = pd.merge(block_gdf,block_top_class1_list[i],on = "block_id",how = "inner")
    block_gdf_classed_list.append(block_gdf_classed)
```
