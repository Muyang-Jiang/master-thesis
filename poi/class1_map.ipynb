## assgin a master ID to all class1

```python
# Function to classify categories into simplified classes
def classify_categories(categories, map):
    classified = {}
    for category in categories:
        class_id = map.get(category, "Unknown")
        if class_id not in classified:
            classified[class_id] = []
        classified[class_id].append(category)
    return classified
```

```python
all_df = pd.read_csv(os.path.join(summary_path, "summary_all.csv"), dtype={0: 'str'})
all_df
```

```python
map_set_list = []
```

### 2022

```python
year  = "2022"
# Original categories in Chinese
original_categories = ast.literal_eval(all_df.loc[all_df['year'] == year]["class1"].values[0])

# Mapping of original categories to new classes with numeric IDs
category_map = {
    '公司企业': 1, '生活服务': 1, '商务住宅': 1, '金融机构': 1, 
    '旅游景点': 2, '休闲娱乐': 2,
    '餐饮美食': 3,
    '购物消费': 4,
    '科教文化': 5,
    '汽车相关': 6, '交通设施': 6,
    '运动健身': 7, '医疗保健': 7,
    '酒店住宿': 8
    
}


# Classify the original categories
classified_categories = classify_categories(original_categories, category_map)

# Print the classified categories with their new IDs
for class_id, categories in classified_categories.items():
    print(f"Class ID {class_id}: {', '.join(categories)}")
    
map_set_list.append(classified_categories)
```

### clean and save map table

```python
index_list = []
class1_list = []
for classified_categories in map_set_list:
    for class_id, categories in classified_categories.items():
        for cat in categories:
            index_list.append(class_id)
            class1_list.append(cat)
```

```python
df_class_map = pd.DataFrame(columns=['class1_label', 'class1'])

df_class_map['class1_label'] = index_list
df_class_map['class1'] = class1_list

df_class_map
```

```python
df_class_map = df_class_map.drop_duplicates()
df_class_map = df_class_map.sort_values(by = "class1_label")
df_class_map.shape[0]
```

```python
df_class_map.to_csv((summary_path+"class1_map.csv"),encoding = "utf-8-sig",index = False)
```
