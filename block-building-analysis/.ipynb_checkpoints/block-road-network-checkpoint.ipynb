```python
import geopandas as gpd
import pandas as pd
import numpy as np
import os
import csv
from tqdm import tqdm
import rasterio

from shapely.geometry import Point
import matplotlib.pyplot as plt
import seaborn as sns
from shapely.geometry import Polygon
import folium

import tifffile
from rasterio.features import rasterize
import seaborn as sns

import tifffile as tiff
from sklearn.metrics import mean_squared_error

import networkx as nx
import osmnx as ox
from shapely.geometry import LineString, Point
from shapely.ops import nearest_points
import time

import fiona
import ast
import statistics
```

```python
accessible_block_save_path = "/data-sat-img/postprocessed/mujiang/master-thesis/accessible_block/"
block_categorized_savepath = "/data-sat-img/RAW/Block/result/block_categorized/"
osmnx_graph_path = "/data-sat-img/postprocessed/mujiang/master-thesis/osmnx_graph/"
```

## read clustered block

```python
file_list = os.listdir(block_categorized_savepath)
```

```python
block_list = []
city_list = []
year = 2020

for file in tqdm(file_list):
    block_file = os.path.join(block_categorized_savepath,file,f"{file}-{year}-block_categorized.shp")
    if os.path.exists(block_file):
        block_list.append(gpd.read_file(block_file))
        city_list.append(file)
```

```python
print("total number of city is", len(block_list))
```

## read road network data in gpkg format

```python
osmnx_graph_nodes_list = []
osmnx_graph_edges_list = []

for city in city_list:
    if os.path.exists(os.path.join(osmnx_graph_path,f"{city}_graph.gpkg")):
        print(city)
        
        osmnx_graph_nodes_list.append(gpd.read_file(os.path.join(osmnx_graph_path,f"{city}_graph.gpkg"),layer = "nodes"))
        osmnx_graph_edges_list.append(gpd.read_file(os.path.join(osmnx_graph_path,f"{city}_graph.gpkg"),layer = "edges"))
```

## road network graph

```python
osmnx_graph_list = []

for city in city_list:
    if os.path.exists(os.path.join(osmnx_graph_path,f"{city}_graph.graphml")):
        print(city)
        osmnx_graph_list.append(ox.load_graphml(os.path.join(osmnx_graph_path,f'{city}_graph.graphml')))
```

## buffer and spatial join

```python
buffer_distance = 10  # meters, adjust based on CRS
search_distance=1500 # search distance in the road network
```

```python
# Pre-compute buffered geometries and join results
buffered_blocks_dict = {}
joined_results_dict = {}

print("buffer block and spatial join")
for i, city in enumerate(city_list):
    if city == "Shenzhen":
        buffered_geometries = block_list[i]['geometry'].buffer(buffer_distance)
        buffered_blocks_gdf = block_list[i].copy()
        buffered_blocks_gdf['geometry'] = buffered_geometries
        buffered_blocks_gdf = buffered_blocks_gdf.to_crs("4326")
        joined_result = gpd.sjoin(osmnx_graph_nodes_list[i], buffered_blocks_gdf, how="inner", predicate='within')
        
        buffered_blocks_dict[i] = buffered_blocks_gdf
        joined_results_dict[i] = joined_result
```

## 用最小值的新方法

```python
# Process each city
for i, city in enumerate(city_list):
    if city == "Shenzhen":
        joined_result = joined_results_dict[i]

        # Create osmid_block_list and block_osmid_list
        osmid_block_list = joined_result[["osmid", "block_id"]].groupby('osmid').agg(list).reset_index()
        block_osmid_list = joined_result[["osmid", "block_id"]].groupby('block_id').agg(list).reset_index()

        # Create accessible blocks
        accessible_blocks_dist_dict_list = []

        # Pre-compute subgraphs for each node
        print("create subgraphs for each node")
        subgraphs = {}
        for osmid in tqdm(osmid_block_list["osmid"]):
            subgraphs[osmid] = nx.ego_graph(osmnx_graph_list[i], osmid, radius=search_distance, distance='length')

        print("find accessbible blocks and calculate distances")
        for index, row in tqdm(block_osmid_list.iterrows(), total=len(block_osmid_list)):
            osmid_list = row["osmid"]
            access_dist_dict = dict()

            for block_osmid in osmid_list:
                subgraph = subgraphs[block_osmid]
                accessible_nodes = subgraph.nodes()

                for accessible_node in accessible_nodes:
                    access_distance = nx.shortest_path_length(osmnx_graph_list[i], source=block_osmid, target=accessible_node, weight='length')
                    accessible_blocks = list(osmid_block_list[osmid_block_list['osmid'].isin([accessible_node])]["block_id"])

                    if accessible_blocks:
                        accessible_blocks = accessible_blocks[0]
                        for accessible_block in accessible_blocks:
                            if accessible_block not in access_dist_dict:
                                access_dist_dict[accessible_block] = []
                            access_dist_dict[accessible_block].append(access_distance)

            if access_dist_dict:
                min_access_dist_dict = {key: min(values) for key, values in access_dist_dict.items()}
                accessible_blocks_dist_dict_list.append(min_access_dist_dict)
            else:
                accessible_blocks_dist_dict_list.append({})

        block_osmid_list["accessible_blocks"] = accessible_blocks_dist_dict_list

        # Save the result
        export_file_name = os.path.join(accessible_block_save_path, str(search_distance), f"{city}-accessible_blocks_min.csv")
        block_osmid_list.to_csv(export_file_name)
        print("Exported result to", export_file_name)
```

## 用平均值的旧方法

```python
# Process each city
for i, city in enumerate(city_list):
    if city == "Shenzhen":
        joined_result = joined_results_dict[i]

        # Create osmid_block_list and block_osmid_list
        osmid_block_list = joined_result[["osmid", "block_id"]].groupby('osmid').agg(list).reset_index()
        block_osmid_list = joined_result[["osmid", "block_id"]].groupby('block_id').agg(list).reset_index()

        # Create accessible blocks
        accessible_blocks_dist_dict_list = []

        # Pre-compute subgraphs for each node
        print("create subgraphs for each node")
        subgraphs = {}
        for osmid in tqdm(osmid_block_list["osmid"]):
            subgraphs[osmid] = nx.ego_graph(osmnx_graph_list[i], osmid, radius=search_distance, distance='length')

        print("find accessbible blocks and calculate distances")
        for index, row in tqdm(block_osmid_list.iterrows(), total=len(block_osmid_list)):
            osmid_list = row["osmid"]
            access_dist_dict = dict()

            for block_osmid in osmid_list:
                subgraph = subgraphs[block_osmid]
                accessible_nodes = subgraph.nodes()

                for accessible_node in accessible_nodes:
                    access_distance = nx.shortest_path_length(osmnx_graph_list[i], source=block_osmid, target=accessible_node, weight='length')
                    accessible_blocks = list(osmid_block_list[osmid_block_list['osmid'].isin([accessible_node])]["block_id"])

                    if accessible_blocks:
                        accessible_blocks = accessible_blocks[0]
                        for accessible_block in accessible_blocks:
                            if accessible_block not in access_dist_dict:
                                access_dist_dict[accessible_block] = []
                            access_dist_dict[accessible_block].append(access_distance)

            if access_dist_dict:
                mean_access_dist_dict = {key: statistics.mean(values) for key, values in access_dist_dict.items()}
                accessible_blocks_dist_dict_list.append(mean_access_dist_dict)
            else:
                accessible_blocks_dist_dict_list.append({})

        block_osmid_list["accessible_blocks"] = accessible_blocks_dist_dict_list

        # Save the result
        export_file_name = os.path.join(accessible_block_save_path, str(search_distance), f"{city}-accessible_blocks.csv")
        block_osmid_list.to_csv(export_file_name)
        print("Exported result to", export_file_name)
```

```python
for i in range(len(city_list)):
    city = city_list[i]
    print(city)
    
    if city == "Shenzhen":
        buffered_geometries = block_list[i]['geometry'].buffer(buffer_distance)

        # Create a new GeoDataFrame that includes the original data with the new buffered geometries
        buffered_blocks_gdf = block_list[i].copy()
        buffered_blocks_gdf['geometry'] = buffered_geometries

        buffered_blocks_gdf = buffered_blocks_gdf.to_crs("4326")
        joined_result = gpd.sjoin(osmnx_graph_nodes_list[i], buffered_blocks_gdf, how="inner", predicate='within')

        # find for each nodes which block they belong too
        osmid_block_list = joined_result[["osmid","block_id"]].groupby('osmid').agg(list).reset_index()

        # find for each block which nodes they covers
        block_osmid_list = joined_result[["osmid","block_id"]].groupby('block_id').agg(list).reset_index()

        # find accessible blocks
        accessible_blocks_dist_dict_list = []
        
        # iterate over blocks
        for index,row in tqdm(block_osmid_list.iterrows(), total=len(block_osmid_list)):
            # the list of nodes are within that block
            osmid_list = row["osmid"]
            access_dist_dict = dict()
            
            # for each node with in the target block,find the accessible nodes within the search distance and record the access disntance
            for block_osmid in osmid_list:
                # create the subgraph for each start point from the sleceted block and get their accessible points
                subgraph = nx.ego_graph(osmnx_graph_list[i], block_osmid, radius=search_distance, distance='length')
                accessible_nodes = subgraph.nodes()               

                # calculate the distance between the origin point from the block and the accessible destination nodes, and find which block does this node belong to
                for accessible_node in accessible_nodes:
                    # origin point from the block and the accessible destination nodes
                    access_distance = nx.shortest_path_length(osmnx_graph_list[i], source=block_osmid, target=accessible_node,weight='length')
                    
                    # for the accessible nodes, find which blocks they belongs to
                    accessible_blocks = list(osmid_block_list[osmid_block_list['osmid'].isin([accessible_node])]["block_id"])
                    if len(accessible_blocks) > 0:
                        accessible_blocks = accessible_blocks[0]

                        for accessible_block in accessible_blocks:
                            if not accessible_block in access_dist_dict:
                                access_dist_dict[accessible_block] = []
                            access_dist_dict[accessible_block].append(access_distance)
                        
            # take the mean value of the distance to acceessible block
            if access_dist_dict:
                mean_access_dist_dict = {key: statistics.mean(values) for key, values in access_dist_dict.items()}
                accessible_blocks_dist_dict_list.append(mean_access_dist_dict)                
                #print(row["block_id"],mean_access_dist_dict)
            else:
                accessible_blocks_dist_dict_list.append({})

        block_osmid_list["accessible_blocks"] = accessible_blocks_dist_dict_list

        # save the result
        export_file_name = os.path.join(accessible_block_save_path,str(search_distance),f"{city}-accessible_blocks.csv")
        block_osmid_list.to_csv(export_file_name)
        print("export result to",export_file_name)
```
